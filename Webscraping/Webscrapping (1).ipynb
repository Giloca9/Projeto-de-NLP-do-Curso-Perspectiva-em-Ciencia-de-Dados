```python

url = "https://www.verdaoweb.com.br/news.asp?nID=158036"

response = requests.get(url)
response.encoding = "utf-8"

soup = BeautifulSoup(response.text, "html.parser")

# titulo
title = soup.find("h1", class_="titulo").get_text(strip=True)

# data
date = soup.find("span", class_="label label-important").get_text(strip=True)

# texto
text = soup.find("div", class_="row-fluid interna").find_all("p")[0].get_text(" ", strip=True)

# visitas
views_text = soup.find("em").get_text(strip=True)
views = int(views_text.split()[0])


print("Título:", title)
print("Data:", date)
print("Texto:", text)
print("Visitas:", views)
```

```python
import time
import random

numero_paginas = 20
links_noticias = []

base_url = "https://www.verdaoweb.com.br/"

for p in range(1, numero_paginas + 1):

    if p == 1:
        url = "https://www.verdaoweb.com.br/noticias2.asp"
    else:
        url = f"https://www.verdaoweb.com.br/noticias2.asp?currentpage={p}"

    response = requests.get(url, headers=headers)
    print(p, response)

    soup = BeautifulSoup(response.text, "html.parser")

    bloco_noticias = soup.find("ul", class_="ultimas-noticias")

    links = [a["href"] for a in bloco_noticias.find_all("a", href=True)]

    links = [
        link if link.startswith("https") else base_url + link
        for link in links
    ]

    links_noticias += links

    tempo_espera = random.uniform(0, 0.01)
    time.sleep(tempo_espera)

print(links_noticias)
print()
print("Total de links:", len(links_noticias))
```

```python
from tqdm import tqdm

dados_noticias = []

# Percorrendo cada link de notícia
for link in tqdm(links_noticias):
    response = requests.get(link, headers=headers)
    response.encoding = "utf-8"

    soup = BeautifulSoup(response.text, "html.parser")

    title = soup.find("h1", class_="titulo").get_text(strip=True)
    date = soup.find("span", class_="label label-important").get_text(strip=True)
    text = soup.find("div", class_="row-fluid interna").find_all("p")[0].get_text(" ", strip=True)
    views_text = soup.find("em").get_text(strip=True)
    views = int(views_text.split()[0])


    noticia = {
        "titulo": title,
        "data": date,
        "texto": text,
        "visitas": views,
    }

    dados_noticias.append(noticia)

    tempo_espera = random.uniform(0, 0.01)
    time.sleep(tempo_espera)
```

```python
import pandas as pd

df_noticias = pd.DataFrame(dados_noticias)

print(df_noticias.head())

df_noticias.to_csv("noticias_palmeiras.csv", index=False, encoding="utf-8-sig", sep=";")
```
